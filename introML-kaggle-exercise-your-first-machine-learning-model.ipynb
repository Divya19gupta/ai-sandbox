## Step 3: Specify and Fit Model
Create a `DecisionTreeRegressor` and save it iowa_model. Ensure you've done the relevant import from sklearn to run this command.

Then fit the model you just created using the data in `X` and `y` that you saved above.

```python
# NOTES:
# DECISION TREE REGRESSOR - QUICK REFERENCE
# What it is:

# A model that makes predictions using yes/no questions (like a flowchart)
# Asks questions about features → Splits data into groups → Predicts average value of that group

# How it works:

# Ask: "Does house have 3+ rooms?" → YES/NO
# Ask: "Does it have 2+ bathrooms?" → YES/NO
# Reach a leaf (bottom) → Predict price = $500,000

# Learning process:

# Model looks at training data
# Finds the BEST questions that separate high-price from low-price houses
# Builds a tree that captures these patterns

# Key terms:

# Leaf = The bottom of the tree where prediction happens
# Split = A yes/no question that divides the data
# Deeper tree = More splits = more specific rules


# RANDOM_STATE - QUICK REFERENCE
# What it is:

# A seed that controls randomness in model building
# Makes results reproducible (same answer every time)

# Why it matters:

# When building a tree, model sometimes has multiple good questions to ask
# Without random_state → picks randomly → different results each run
# With random_state=1 → always picks the SAME way → same results each run

# Example:
# Without random_state: [1035000, 1465000, ...] then [1034000, 1466000, ...] ← DIFFERENT
# With random_state=1: [1035000, 1465000, ...] then [1035000, 1465000, ...] ← SAME
# Number doesn't matter: Use any number (1, 42, 999) — just be consistent!

# CODE:
# from _ import _
from sklearn.tree import DecisionTreeRegressor
#specify the model. 
#For model reproducibility, set a numeric value for random_state when specifying the model
iowa_model = DecisionTreeRegressor(random_state=1)

# Fit the model
iowa_model.fit(X,y)

# Check your answer
step_3.check()
```

## Step 4: Make Predictions
Make predictions with the model's `predict` command using `X` as the data. Save the results to a variable called `predictions`.
